```python
import numpy as np
```

```python
def read_input():
    with open('day4-1.txt') as f:
        return f.read()
```

```python
a=read_input()
a=a.split('\n')
for i in range(0,len(a)):
    a[i]=a[i].split(' ')
a=np.array(a)
```

```python
a
```

```python
ans=np.zeros(len(a))
for i in range(a.shape[0]):
    for j in range(len(a[i])):
        for k in range(j)+range(j+1,len(a[i])):
            if a[i][j]==a[i][k]:
                ans[i]=ans[i]+1
np.count_nonzero(ans)
```
